```python
data = data.drop(['geographic name, french','georaphie type, english', 'georaphie type, french','census division code', 'economic region code', 'canadian heritage region code'],axis=1)
data
```

```python
def is_int(x):
    try:
        int(float(x.replace(",","")))
    except ValueError:
        return False
    return True

data = data.replace('([0-9]),([0-9])','\\1\\2', regex=True)#Remove all commas surrounded by numbers
for i in ['English first official language spoken, adjusted',
       'French first official language spoken, adjusted',
       'Other first official language spoken, adjusted']:
    data = data.drop(data[data[i].apply(lambda x: not is_int(x))].index)
    data[i] = data[i].astype(float)
    data[i] = data[i].astype(int)
data.dtypes
```

```python
#Search "Province code" https://www150.statcan.gc.ca/n1/pub/92-195-x/2011001/geo/prov/tbl/tbl8-eng.htm
data = data.replace(['pr10','pr11','pr12','pr13','pr24','pr35','pr46','pr47','pr48','pr59','pr60','pr61','pr62','n.d.'],['NL','PE','NS','NB','QC','ON','MB','SK','AB','BC','YT','NT','NU','ND'])

```

```python
division = data[data['geographic name, english'].str.contains("^Division*")]
data['geographic name, english'] = data.apply(lambda x: x['geographic name, english'].replace('Division',x['province code']), axis=1)
data = data.set_index('geographic name, english')
division ['geographic name, english'] = division.apply(lambda x: x['geographic name, english'].replace('Division',x['province code']), axis=1)
division = division.set_index('geographic name, english')
division
```

```python
def  plot_province(df,province,english=False,french=True,other=True,setProvCode=False):
    df = df[df['province code']== province.upper()]
    toplot = []
    if english:
        toplot.append('English first official language spoken, adjusted')
    if french:
        toplot.append('French first official language spoken, adjusted')
    if other:
        toplot.append('Other first official language spoken, adjusted')
    df[toplot].plot(kind="bar",figsize=(20,10),ylabel="Speakers")
plot_province(division,"MB")
```

```python
totals = pd.DataFrame().reindex_like(data)
totals = totals[0:0]
def get_totals_province(df,province):
    numbers = df[df['province code']== province.upper()]
    return sum(numbers['English first official language spoken, adjusted']), sum(numbers['French first official language spoken, adjusted']), sum(numbers['Other first official language spoken, adjusted'])

totals
```

```python
province_codes = ['NL','PE','NS','NB','QC','ON','MB','SK','AB','BC','YT','NT','NU']
for i in province_codes:
    r = get_totals_province(data,i)
    columns_dic = {"province code":i,
                    "Total":sum(r),
                    "English first official language spoken, adjusted":r[0],
                    "French first official language spoken, adjusted":r[1],
                    "Other first official language spoken, adjusted":r[2]}

    totals =totals.append(columns_dic,ignore_index=True)
totals
```
